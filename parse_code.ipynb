```python
import re, os
```

## Read CHAR FILE

```python
char_file_path = 'files_hexa/CHARC_PROMED_20240613_02-36.log.txt'
with open(char_file_path, 'r', encoding='utf-8') as f:
    char_content = f.read()
```

```python
pattern_tool = r'[A-Z]\|\d+\|[A-Z]+\|[0-9-]+\^([A-Za-z\(\)%#*\-]+)\^\w+\|\|([0-9.]+)'
pattern_name = r'PID\|\d+\|\|[^\|]*\|[^\|]*\|\^([A-Za-z\ \']+)'
```

```python
matches = re.findall(pattern=pattern_tool, string=char_content)
name = re.findall(pattern=pattern_name, string=char_content)
```

```python
matches.pop(0)
print(matches)
print(name)
```

```python
print(f'Name: {name[0]}')
for match in matches:
    parameter, value = match
    print(f'Parameter: {parameter}, Nilai: {value}')
```
